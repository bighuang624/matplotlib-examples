```python
RRSE_df = pd.read_excel('final_ablation_study_RRSE.xlsx')
MAE_df = pd.read_excel('final_ablation_study_MAE.xlsx')
```

```python
fig1, (ax1) = plt.subplots(1, 1, figsize=(6,5))
sns.barplot(x="window-horizon", y="RRSE",
                hue="model", data=RRSE_df, ax=ax1)
plt.ylim(0.65, 1.05)    # 设置纵坐标范围
plt.tick_params(labelsize=14)    # 设置坐标刻度字体大小
legend = plt.legend(loc = 4, title=None, framealpha=1, fontsize='x-large')
ax1.set_xlabel(xlabel='window-horizon', fontsize=18)
ax1.set_ylabel(ylabel='RRSE',fontsize=18)
plt.tight_layout()    # 避免标签超出显示范围
plt.savefig('final_ablation_RRSE.eps', dpi=600, format='eps')
```

```python
fig2, (ax2) = plt.subplots(1, 1, figsize=(6,5))
sns.barplot(x="window-horizon", y="MAE",
                hue="model", data=MAE_df, ax=ax2)
plt.ylim(0.32, 0.56)
plt.tick_params(labelsize=14)
plt.legend(loc = 4, title=None, framealpha=1, fontsize='x-large')
ax2.set_xlabel(xlabel='window-horizon', fontsize=18)
ax2.set_ylabel(ylabel='MAE',fontsize=18)
plt.tight_layout()
plt.savefig('final_ablation_MAE.eps', dpi=600, format='eps')
```
