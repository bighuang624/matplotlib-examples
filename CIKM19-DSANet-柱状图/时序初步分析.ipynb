```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")
```

```python
file_path = 'wecar.txt'
raw_data = np.loadtxt(open(file_path), delimiter=',')
```

```python
df = pd.DataFrame(raw_data, columns=['var1', 'var2', 'var3', 'var4', 'var5'])
```

```python
# multivariate plot
fig1, (ax1) = plt.subplots(1, 1, figsize=(15, 8))
sns.lineplot(data=df, linewidth=1.5, ax=ax1, dashes=False)
ax1.set_xlabel(xlabel='time', fontsize=14)
ax1.set_ylabel(ylabel='value',fontsize=14)
# plt.savefig('multivariate_plot.eps', dpi=650, format='eps')
```

```python
# univariate plot
fig1, (ax1) = plt.subplots(1, 1, figsize=(15, 8))
ax = sns.lineplot(data=df["var1"], color="coral", label="var1", ax=ax1)
```

## Autocorrelation

```python
time_vary = 900
autocorr_1 = [df["var1"].autocorr(lag=i) for i in range(time_vary)]
autocorr_2 = [df["var2"].autocorr(lag=i) for i in range(time_vary)]
autocorr_3 = [df["var3"].autocorr(lag=i) for i in range(time_vary)]
autocorr_4 = [df["var4"].autocorr(lag=i) for i in range(time_vary)]
autocorr_5 = [df["var5"].autocorr(lag=i) for i in range(time_vary)]

autocorr = {'var1': autocorr_1, 'var2': autocorr_2, 'var3': autocorr_3, 'var4': autocorr_4, 'var5': autocorr_5}
```

```python
# autocorr_df = pd.DataFrame(autocorr, columns=['var1', 'var2', 'var3', 'var4', 'var5'])
autocorr_df = pd.DataFrame(autocorr)
```

```python
fig1, (ax1) = plt.subplots(1, 1, figsize=(12, 6))
sns.lineplot(data=autocorr_df, color="coral", ax=ax1, dashes=False, legend=False)
plt.xlim(0, time_vary+20)
ax1.set_xlabel(xlabel='Time Lag (days)', fontsize=14)
ax1.set_ylabel(ylabel='Autocorrelation', fontsize=14)
plt.savefig('autocorr.eps', dpi=650, format='eps')
```
